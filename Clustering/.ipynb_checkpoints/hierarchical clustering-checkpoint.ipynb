```python
import numpy as np
from sklearn.cluster import MeanShift #as ms <- didn't work
from sklearn.datasets.samples_generator import make_blobs #data
import matplotlib.pyplot as plt
```

```python
centers = [[1,1], [5,5]]
X, _ = make_blobs(n_samples = 200, centers = centers, cluster_std = 1) #_ underscore means its useless
#std standard divation
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show
```

```python
ms = MeanShift()
ms.fit(X)
labels=ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
n_clusters_ = len(np.unique(labels))
print(f"Numbers of estimated clusters: {n_clusters_}")

colors = 10*['r.','g.','b.','c.','m.','y.','k.']

print(colors)
print(labels)
```

```python
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)

plt.scatter(cluster_centers[:,0],cluster_centers[:,1],
            marker="x",color='k', s=150, linewidths = 5, zorder=10)

plt.show()
```
