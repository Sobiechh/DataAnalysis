```python
import matplotlib.pyplot as plt
```

```python
slices = [5,10,52,33]
cars = ["Dacia", "Volvo", "BMW", "Audi"]
cols = ['r', 'grey', 'blue', 'y']
```

```python
plt.pie(slices, 
        labels = cars, 
        colors = cols,
        startangle = 90,
        shadow = True,
        explode=(0.01,0.01,0.01,0.01),
        autopct ="%1.1f%%")
plt.title("Cars in Poland")
plt.show()
```
