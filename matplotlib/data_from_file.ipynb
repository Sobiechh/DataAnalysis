```python
import matplotlib.pyplot as plt
import csv
```

```python
x = []
y = []
```

```python
with open('data.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
```

```python
plt.plot(x,y, label = "hours working")
plt.xlabel('x')
plt.ylabel('y')
plt.title("loaded from file")
plt.legend()
plt.show()
```
